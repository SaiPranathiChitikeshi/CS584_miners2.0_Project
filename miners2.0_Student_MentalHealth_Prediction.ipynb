```python
BMIbin = []
for i in student_data_and_HADS.itertuples(): 
    if(i[13]<=16):
        BMIbin.append(0)
    elif(i[13]<18.5):
        BMIbin.append(1)
    elif(i[13]<25):
        BMIbin.append(2)
    elif(i[13]<30):
        BMIbin.append(3)
    else:
        BMIbin.append(4)
BMIbin

```

```python
currEdu = []
for i in student_data_and_HADS.itertuples(): 
    if(i[3] == 'G'):
        currEdu.append(1)
    elif(i[3] == 'PG'):
        currEdu.append(2)
    else:
        currEdu.append(0)
currEdu
```

```python
Exercise = []
for i in student_data_and_HADS.itertuples(): 
    if(i[4] == 'None'):
        Exercise.append(0)
    elif(i[4] == 'Heavy'):
        Exercise.append(2)
    else:
        Exercise.append(1)
Exercise
```

```python
student_data_and_HADS['BMIbin'] = BMIbin
student_data_and_HADS['Current Education'] = currEdu
student_data_and_HADS['Exercise'] = Exercise
student_data_and_HADS.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
student_data_and_HADS.drop(['StudentID'],axis = 1,inplace = True)
```

```python
scaler = StandardScaler()
standadized_dataFrame = scaler.fit_transform(student_data_and_HADS.drop(['HADSbin'],axis = 1))
standadized_dataFrame = pd.DataFrame(standadized_dataFrame)
standadized_dataFrame
```

```python
student_data_and_HADS.columns
```

```python
standadized_dataFrame.columns = ['IQScore', 'Current Education', 'Exercise', 'Gender', 'Height',
       'Weight', 'Age', 'Peer Pressure', 'Living with Family', 'Past History',
       'Smoke', 'BMIbin']
standadized_dataFrame.head()
```

```python
student_data_and_HADS.info()
```

```python
suicide_possibility = []
for i in student_data_and_HADS.itertuples(): 
    if(i[12] == 'Abnormal'):
        suicide_possibility.append('Yes')
    elif(i[12] == 'Normal'):
        suicide_possibility.append('No')
    else:
        suicide_possibility.append('Maybe')
suicide_possibility
```

```python
standadized_dataFrame['Suicide Possibility'] = suicide_possibility
```

```python
X = standadized_dataFrame.drop(['Suicide Possibility'],axis = 1)
y = standadized_dataFrame['Suicide Possibility']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
print('Accuracy: ',accuracy_score(y_test, y_predict))
print("Precision Score : ",precision_score(y_test, y_predict, average='micro'))
print("Recall Score : ",recall_score(y_test, y_predict, average='micro'))
print('F1 score: ', f1_score(y_test, y_predict, average='micro'))
```

```python
model = KNeighborsClassifier()
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
print('Accuracy: ',accuracy_score(y_test, y_predict))
print("Precision Score : ",precision_score(y_test, y_predict, average='micro'))
print("Recall Score : ",recall_score(y_test, y_predict, average='micro'))
print('F1 score: ', f1_score(y_test, y_predict, average='micro'))
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
print('Accuracy: ',accuracy_score(y_test, y_predict))
print("Precision Score : ",precision_score(y_test, y_predict, average='micro'))
print("Recall Score : ",recall_score(y_test, y_predict, average='micro'))
print('F1 score: ', f1_score(y_test, y_predict, average='micro'))
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
print('Accuracy: ',accuracy_score(y_test, y_predict))
print("Precision Score : ",precision_score(y_test, y_predict, average='micro'))
print("Recall Score : ",recall_score(y_test, y_predict, average='micro'))
print('F1 score: ', f1_score(y_test, y_predict, average='micro'))
```
